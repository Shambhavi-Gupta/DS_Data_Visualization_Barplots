```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use(['dark_background'])
import seaborn as sns
sns.set(color_codes=True)
```

```
t = sns.load_dataset('tips')
```

```
sns.barplot(x ='day' ,y ='tip' ,data =t);
```

```
sns.barplot(x ='day' ,y ='tip' ,data =t ,estimator =np.median);
```

```
def my_estimate(v):
    return np.quantile(v ,0.25)
```

```
sns.barplot(x ='day' ,y ='tip' ,data =t ,estimator =my_estimate);
```

```
sns.barplot(x ='day' ,y ='tip' ,data =t , hue ='sex', estimator =np.median);
```

```
sns.barplot(x ='day' ,y ='tip' ,data =t , hue ='smoker', estimator =np.median);
```

```
sns.barplot(x ='day' ,y ='tip' ,data =t , hue ='time', estimator =np.median);
```
